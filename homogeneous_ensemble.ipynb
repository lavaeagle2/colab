```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

flower_data = {
    'Petal_Color': ['Red', 'Pink', 'Red', 'Yellow', 'Red'],
    'Petal_Size': [5.2, 2.1, 1.8, 4.5, 5.0],
    'Target_Label': ['Rose', 'Hibiscus', 'Rose', 'Sunflower', 'Rose']
}
df = pd.DataFrame(flower_data)
le_color = LabelEncoder()
df['Color_Enc'] = le_color.fit_transform(df['Petal_Color'])
le_target = LabelEncoder()
y = le_target.fit_transform(df['Target_Label'])
X = df[['Color_Enc', 'Petal_Size']]

model_a = DecisionTreeClassifier(max_depth=1, random_state=42)
model_b = DecisionTreeClassifier(max_depth=5, random_state=42)

model_a.fit(X, y)
model_b.fit(X, y)

pred_a = model_a.predict(X)
pred_b = model_b.predict(X)

ensemble_preds = []
for a, b in zip(pred_a, pred_b):
    ensemble_preds.append(a if a == b else b)

results = pd.DataFrame({
    'Model_A': le_target.inverse_transform(pred_a),
    'Model_B': le_target.inverse_transform(pred_b),
    'Ensemble': le_target.inverse_transform(ensemble_preds)
})
print(results)
```
