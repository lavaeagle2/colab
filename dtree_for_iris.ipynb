```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```
data = load_iris()
df = pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```
print(df.shape)
print('')
print(df.isnull().sum())
```

```
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
```

```
acc_tree = accuracy_score(y_test, tree.predict(X_test))
acc_tree
```

```
rf = RandomForestClassifier(n_estimators=5, random_state=42)
rf.fit(X_train, y_train)
```

```
rf1 = RandomForestClassifier(n_estimators=10, random_state=42)
rf1.fit(X_train, y_train)
```

```
rf2 = RandomForestClassifier(n_estimators=20, random_state=42)
rf2.fit(X_train, y_train)
```

```
rf3 = RandomForestClassifier(n_estimators=40, random_state=42)
rf3.fit(X_train, y_train)
```

```
rf4 = RandomForestClassifier(n_estimators=70, random_state=42)
rf4.fit(X_train, y_train)
```

```
acc_rf = accuracy_score(y_test, rf.predict(X_test))
acc_rf
```

```
acc_rf = accuracy_score(y_test, rf1.predict(X_test))
acc_rf
```

```
acc_rf = accuracy_score(y_test, rf2.predict(X_test))
acc_rf
```

```
acc_rf = accuracy_score(y_test, rf3.predict(X_test))
acc_rf
```

```
acc_rf = accuracy_score(y_test, rf4.predict(X_test))
acc_rf
```

```
trees = [1,5,10,50,100]
accs = []
for n in trees:
    model = RandomForestClassifier(n_estimators=n, random_state=42)
    model.fit(X_train, y_train)
    accs.append(accuracy_score(y_test, model.predict(X_test)))

plt.figure()
plt.plot(trees, accs, marker='o')
plt.title('Accuracy vs Trees')
plt.show()
```
