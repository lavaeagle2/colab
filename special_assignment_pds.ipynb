```python
import pandas as pd
df= pd.read_csv('/content/Titanic-Dataset.csv')
df
```

1. Consider the Titanic Dataset, plot the variation of surviving passengers v/s their age. Also, show their gender (male/female).

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x='Age', y='Survived', hue='Sex', data=df)
plt.title('Survival Status vs. Age by Gender')
plt.xlabel('Age')
plt.ylabel('Survived')
plt.yticks([0, 1],['no','yes'])
plt.show()
```

2. With the Titanic Dataset, plot the variation of Fare v/s passenger class. Also, show their gender
(male/female) and age in the same plot


```python
sns.scatterplot(x='Fare', y='Pclass', hue='Sex',size='Age', data=df)
plt.title('Fare vs. passenger class')
plt.xlabel('Fare')
plt.ylabel('Passengar class')
plt.yticks([0, 1,2,3])
plt.show()
```

3. Plot the survival rate of male and female passengers v/s their passenger class

```python
survival_by_pclass = df.groupby(['Pclass','Sex'])['Survived'].mean().reset_index()
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=survival_by_pclass)
plt.title('Survival Status vs. Passenger Class by Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survived')
plt.yticks([0, 1], ['No', 'Yes'])
plt.show()
```

4. what sorts of people were more likely to survive?” using passenger data (ie name, age, gender, socio-economic class, etc)?

```python
survival_by_sex = df.groupby('Sex')['Survived'].mean().reset_index()
print(survival_by_sex)
print('')
plt.figure(figsize=(6, 4))
sns.barplot(x='Sex', y='Survived', hue='Sex', data=survival_by_sex)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')
plt.ylim(0, 1)
plt.show()
```

```python
survival_by_pclass = df.groupby('Pclass')['Survived'].mean().reset_index()
print(survival_by_pclass)
print('')

plt.figure(figsize=(7, 5))
sns.barplot(x='Pclass', y='Survived', hue='Pclass', data=survival_by_pclass)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.ylim(0, 1)
plt.show()
```

```python
survival_by_sex_pclass = df.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack()
print(survival_by_sex_pclass)
print('')
survival_by_sex_pclass.plot(kind='bar', figsize=(10, 6))
plt.title('Survival Rate by Passenger Class and Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.show()
```
