```python
# pandas - critical library - data
# csv files, rows and columns, tabular data
# Series & DataFrame - 1D, & 2D
# Series - 1D - single column - a single columns
# DataFrame - 2D - multiple columns, 2D - a table

import pandas as pd

alist = [0, 1, 12, 123, 1234, 11]
print(alist)

aseries = pd.Series(alist, index=['a', 'b', 'c', 'd', 'e', 'f'])
print(aseries)

# on top of index, pandas uses a label
```

```python
# creating series directly from dict

adict = {
    'a':0,'b':1,'c':12,'d':123,'e':1234,'f':11
}
print(adict)

anotherseries = pd.Series(adict)
print(anotherseries)
```
