EDA and Cleaning

```python
import pandas as pd
df = pd.read_csv('/content/E-commerce Customer Behavior - Sheet1.csv')
df
```

```python
df.describe()
```

```python
df = df.drop(['City'], errors='ignore')
df
```

Plotting spend

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.hist(df['Total Spend'],color = 'violet',edgecolor='black',bins=range(400,1600,100))
# sns.histplot(df['Total spend'], binwidth=100)
plt.xlabel('Total spend')
plt.ylabel('Frequency')
plt.title('Distribution of Total spend')
plt.show()
```

majority of people spend between 400-500 rupees while mean customers also spend at any feasible range expect 900-1100


plotting category wise spend

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Membership Type', y='Total Spend', data=df)

plt.title("Category vs Spend Distribution")
plt.xlabel("Category")
plt.ylabel("Spend")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.box(df, x="Membership Type", y="Total Spend", title="Category vs Spend distribution")
fig.show()
```

people with gold membership spends the most where as bronze spent the least

distribution of average item purchased per category and gender

```python
avg_items_purchased = df.groupby(['Gender', 'Membership Type'])['Items Purchased'].mean().reset_index()

plt.figure(figsize=(10,6))

sns.barplot(
    x='Membership Type',
    y='Items Purchased',
    hue='Gender',
    data=avg_items_purchased
)

plt.title("Average Items Purchased by Gender and Membership Type")
plt.xlabel("Membership Type")
plt.ylabel("Average Items Purchased")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

the frequency of items purchased increases with the membership like gold purchses the highest avg items of all and bronze the lowest average of all

male buyers spent more and purchase more also with higher memberships , while females all almost uncertain in their purchases within all kind of categories

Are customers happy with disount?

```python
plt.figure(figsize=(9,6))
sns.scatterplot(
    x='Average Rating',
    y='Discount Applied',
    data=df,
    s=70,
    hue='Gender',
    edgecolor='black'
)

plt.title("Average Rating vs Discount Applied", fontsize=14, weight='bold')
plt.xlabel("Average Rating", fontsize=12)
plt.ylabel("Discount Applied", fontsize=12)
plt.yticks([0, 1], ['No', 'Yes'])

plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

majority of men are still happy without discount where as most of the womens are still unsatisfied after having a discount
